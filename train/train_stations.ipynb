```python
points_geometry = train_stations['geometry']
lon = []
lat = []
for i in range(len(points_geometry)):
    x = points_geometry[i].coords[0][0]
    y = points_geometry[i].coords[0][1]
    x = float(x)
    y = float(y)
    lon.append(x)
    lat.append(y)
```

```python
stops = train_stations[['OBJECTID_l','STN_NAME','PT_CODE1','SUBZONE_N']]
pts = []
for x,y in zip(lon,lat):
    pts.append(Point(x,y))
    
mrt_stations = gpd.GeoDataFrame(stops, geometry=pts, crs="EPSG:4326")
mrt_stations
```
