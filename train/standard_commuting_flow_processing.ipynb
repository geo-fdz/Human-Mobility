```python
def get_raw_data(path,i):
    lis_file = os.listdir(path=path)
    
    df = pd.read_csv(path+'\\'+lis_file[i])
    
    return df
```

```python
szone_tstations_matching = pd.read_csv(r"D:\NUS\project\LTA_data\point_data\train_stations\mrt_matching_LTA.csv")
lis_ts_n = list(set(szone_tstations_matching['PT_CODE1']))
print('The total number of train stations: {}'.format(len(lis_ts_n)))
```

```python
def aggreg_od(od):
    origin,destination,trips_num = 'ORIGIN_PT_CODE','DESTINATION_PT_CODE','TOTAL_TRIPS'
    df = od.groupby([origin,destination]).sum(trips_num).reset_index().drop('TIME_PER_HOUR',axis = 1)
    df['Origin_Subzone'] = None
    df['Destination_Subzone'] = None
    return df
```

```python
def output_subzone(x,df):
    stops =  df.loc[df['PT_CODE1'] == x].copy()
    stops.index = range(len(stops))
    subzone_n = stops.loc[:,'SUBZONE_N'].copy()
    subzone_n = subzone_n.values[0]
    return subzone_n
   
def origin_bus_stops_subzone_matching(subzone_n,x,aggregate_od):
    df_origin = aggregate_od.loc[aggregate_od['ORIGIN_PT_CODE'] == x].copy()
    df_origin.loc[:,'Origin_Subzone'] = subzone_n
    return df_origin

def destination_bus_stops_subzone_matching(subzone_n,x,aggregate_od):
    df_destination = aggregate_od.loc[aggregate_od['DESTINATION_PT_CODE'] == x].copy()
    df_destination.loc[:,'Destination_Subzone'] = subzone_n
    return df_destination
```

```python
def get_origin_subzone(aggregate_od, lis_station, matched_name):
    cols = aggregate_od.columns
    tmp_origin = pd.DataFrame(columns = cols)
    for i in list(lis_station):
        subzone_n = output_subzone(i,matched_name)
        subzone_mrt_stop = origin_bus_stops_subzone_matching(subzone_n,i,aggregate_od)
        tmp_origin = tmp_origin.append(subzone_mrt_stop)
        del(subzone_n)
    return tmp_origin
```

```python
def get_destination_subzone(aggregate_od, lis_station, matched_name):
    cols = aggregate_od.columns
    tmp_destination = pd.DataFrame(columns = cols)
    for i in list(lis_station):
        subzone_n = output_subzone(i,matched_name)
        subzone_mrt_stop = destination_bus_stops_subzone_matching(subzone_n,i,aggregate_od)
        tmp_destination = tmp_destination.append(subzone_mrt_stop)
        del(subzone_n)
    return tmp_destination
```

```python
def main():
    
    raw_data = get_raw_data(path,1)
    aggregate_train_od = aggreg_od(raw_data)
    lis_file = os.listdir(path=path)
    print(lis_file[2])   
    
    train_origin_subzone = get_origin_subzone(aggregate_train_od, lis_ts_n, szone_tstations_matching)
    train_destination_subzone = get_destination_subzone(aggregate_train_od, lis_ts_n, szone_tstations_matching)
    
    train_origin_subzone['index'] = train_origin_subzone.index
    train_destination_subzone['index'] = train_destination_subzone.index
    
    train_origin_subzone = train_origin_subzone.sort_values('index')
    train_destination_subzone = train_destination_subzone.sort_values('index')
    
    train_origin_subzone['Destination_Subzone'] = train_destination_subzone['Destination_Subzone']
    train_od_subzone = train_origin_subzone.copy()
    
    train_od_subzone = train_od_subzone.drop('index',axis = 1)
    
    return train_od_subzone


if __name__ == '__main__':
    train_od_subzone = main()
```
